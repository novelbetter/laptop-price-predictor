```python
#extract x and y resolution

def Xresolution(x_d):
    return x_d.split()[-1].split("x")[0]

def Yresolution(y_d):
    return y_d.split()[-1].split("x")[1]

#find x and y
data['X_res'] = data['ScreenResolution'].apply(lambda x: Xresolution(x))
data['Y_res'] = data['ScreenResolution'].apply(lambda y: Yresolution(y))

#convert to numeric
data['X_res'] = data['X_res'].astype('int')
data['Y_res'] = data['Y_res'].astype('int')

data.info()
data.head()
```

```python
#cpu column extraction
def processor_type(text):
    cpu = " ".join(text.split()[0:3])
    if cpu == 'Intel Core i7' or cpu == 'Intel Core i5' or cpu == 'Intel Core i3' :
        return cpu
    elif cpu.split()[0] == 'Intel':
        return 'Other Intel Processor'
    else:
        return 'AMD Processor'
    
data['CPU_Name'] = data['Cpu'].apply(lambda x: processor_type(x))
data.sample(7)
```

```python
#4 most common observed: ssd, hdd, flash, hybrid

#remove decimal such as 1.0 -> 1
data['Memory'] = data['Memory'].astype(str).replace('\.0','',regex=True)

#replace gb worh with ''
data['Memory'] = data['Memory'].str.replace('GB', '')

#replace TB word with '000'
data['Memory'] = data['Memory'].str.replace('TB', '000')

#split the word accross '+' character
new_data = data['Memory'].str.split("+", n=1, expand=True)

new_data
```

```python
#input column 0 from new_data to data
data['first'] = new_data[0]
data['first'] = data['first'].str.strip
data.head()
```

```python
#flag for disk type
def applychanges(value):
    data['Layer1'+value] = data['first'].astype(str).apply(lambda x:1 if value in x else 0)

listtoapply = ['HDD', 'SSD', 'Flash Storage', 'Hybrid']
for value in listtoapply:
    applychanges(value)

data.head()
```

```python
#flag for disk type
def applychanges1(value):
    data['Layer2'+value] = data['second'].astype(str).apply(lambda x:1 if value in x else 0)

listtoapply = ['HDD', 'SSD', 'Flash Storage', 'Hybrid']
for value in listtoapply:
    applychanges1(value)

#remove character and keep numbers
data['first'] = data['first'].str.replace(r'\D','')
data['first'].value_counts()

data.head()
```

```python
data['first'] = pd.to_numeric(data['first'], errors='coerce')
data['second'] = pd.to_numeric(data['second'], errors='coerce')

data['first'] = data['first'].astype('Int64')
data['second'] = data['second'].astype('Int64')
data.head()
```

```python
#multiplying elements and storing result in subsequent columns

data['HDD'] = (data['first']*data['Layer1HDD']+data['second']*data['Layer2HDD'])
data['SSD'] = (data['first']*data['Layer1SSD']+data['second']*data['Layer2SSD'])
data['Hybrid'] = (data['first']*data['Layer1Hybrid']+data['second']*data['Layer2Hybrid'])
data['Flash_Storage'] = (data['first']*data['Layer1Flash Storage']+data['second']*data['Layer2Flash Storage'])

#drop unnecessary column
data.drop(columns=['first','second','Layer1HDD','Layer1SSD','Layer1Hybrid','Layer1Flash Storage','Layer2HDD','Layer2SSD','Layer2Hybrid','Layer2Flash Storage'])
data.sample(5)
```

```python
#model building
from sklearn.model_selection import train_test_split #menguji besarnya data taining test
from sklearn.pipeline import Pipeline #assemble gabung bbrp step cross validated
from sklearn.compose import ColumnTransformer #apply transformer data heterogen
from sklearn.preprocessing import OneHotEncoder 
from sklearn import metrics #menghitung r square/determinasi pd func regresi
from sklearn.ensemble import RandomForestRegressor
```

```python
#split train and test setx
X= data.drop(columns=['Price'])
y= np.log(data['Price'])

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15, random_state=2)
X.head()
```

```python
X_train.shape, X_test.shape
```

```python
mapper = {i:value for i, value in enumerate(X_train.columns) }
mapper
```

```python
#pipeline

step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,3,8,12])
])

step2 = RandomForestRegressor(
    n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15)

pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

#training model
pipe.fit(X_test,y_train)
#prediksi
y_pred = pipe.predict(X_test)
#evaluasi
y_pred('R2 Score: ' , metrics.r2_score(y_test, y_pred))
print('MAE: ', metrics.mean_absolute_error(y_test,y_pred))
```
